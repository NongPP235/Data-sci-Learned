# DataFrame

## DataFrame from `List`

```python
# DataFrame from a single list
datas = [25, 22, 25, 35, 32]
df = pd.DataFrame(datas)
df
```

```python
# DataFrame from a single list + include "column"
datas = [25, 22, 25, 35, 32]
cols = ['Age']

df = pd.DataFrame(datas, columns=cols)

print(type(df))
df
```

```python
# DataFrame from multiple list (matrix)
datas = [[163, 37, 1],
        [175, 20, 0],
        [162, 27, 0],
        [150, 10, 1] ]

colms = ['height','hair_length','voice_pitch']

df = pd.DataFrame(datas, columns=colms,index=[1,2,3,4])
df
```

```python
# DataFrame from multiple list (matrix)
datas = [['Kobkiat',25,35],
         ['ณิแนน',22,26.0],
         ['จอห์น',25,32],
         ['Tim',35,29],
         ['Steal',32,27]]

cols = ['Name','Age','Score']

df = pd.DataFrame(datas, columns=cols)
df
```

## DataFrame from `Dictionary`

```python
# DataFrame from Dictionary

datas = [{'Name':'Kobkiat','Age':25,'Score':35},
         {'Name':'ณิแนน','Age':22,'Score':26},
         {'Name':'จอห์น','Age':25,'Score':32},
         {'Name':'Tim','Age':35,'Score':29},
         {'Name':'Steal','Age':32,'Score':27}]

df = pd.DataFrame(datas, columns=cols)
df
```

## DataFrame from `Series`

```python
names = ['Kobkiat','ณิแนน','จอห์น','Tim','Steal']
ages = [25, 22, 25, 35, 32]
scores = [35, 26, 32, 29, 27]
cols = ['Name','Age','Score']

name_series = pd.Series(names)
age_series = pd.Series(ages)
score_series = pd.Series(scores)

frame = {'Name':name_series,
         'Age':age_series,
         'Score':score_series}

df = pd.DataFrame(frame)
df
```
