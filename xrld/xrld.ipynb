```python
wb = xl.Workbook()
ws = wb.worksheets[0]

data = np.random.randint(1000,9999,10)
row = 1

for i in data:
    ws.cell(row,1).value = i

    if i >= 5000:
        ws.cell(row,2).value = 'Most Value'
    else:
        ws.cell(row,2).value = 'Less Value'

    row += 1

wb.save('data.xlsx')
```
